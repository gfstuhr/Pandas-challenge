```python
#Importing dependancies and file
import pandas as pd
file_path=("Resources/purchase_data.csv")
purchase_df=pd.read_csv(file_path)
```

```python
#View original csv
purchase_df.head()
```

```python
#Player count
player_count=len(purchase_df["SN"].value_counts())
pc_table = pd.DataFrame([{"Total Player Count": player_count}])
pc_table
```

```python
#Purchasing Analysis Total
#number of unique items
unique_items = len(purchase_df["Item ID"].value_counts())
#average purchase price
avg_pur_price = round(purchase_df["Price"].mean(),2)
#total purchases
total_purchases = purchase_df["Purchase ID"].count()
#total purchase value
total_value = round((purchase_df["Price"].sum()),2)
#creating dataframe of values
purchasing_analysis_total = pd.DataFrame([{
    "Total Purchases": total_purchases,"Number of Unique Itmes": unique_items,
    "Average Purchase Price":avg_pur_price,"Total Revenue":total_value
}])
#Formatting values
purchasing_analysis_total["Average Purchase Price"] = (
purchasing_analysis_total["Average Purchase Price"].astype(float).map("${:,.2f}".format))
purchasing_analysis_total["Total Revenue"] = (
purchasing_analysis_total["Total Revenue"].astype(float).map("${:,.2f}".format))
purchasing_analysis_total
```

```python
#Gender Demographics
#Gender Counts
gender_group = purchase_df.groupby(["Gender"])
gender_unique = gender_group["SN"].value_counts()
gender_count = gender_unique.count("Gender")
#Percent_gender
percent_gender = round((gender_count/gender_unique.count())*100,2)
#Table and Formatting
g_table = pd.DataFrame({
    "Total Gender Count":gender_count,"Percent of Gender":percent_gender
})
g_table = g_table.sort_values("Percent of Gender", ascending=False)
g_table["Percent of Gender"] = g_table["Percent of Gender"].astype(float).map("{:,.2f}%".format)
g_table
```

```python
#Purchasing Analysis Gender
gender_group.head()
```

```python
#Purchasing Analysis Gender
#Counts per gender
purchase_count=gender_group["Purchase ID"].count()
#Avg purchase price
avg_pur_price_gender = round(gender_group["Price"].mean(),2)
#Total purchase value
total_value_gender = round(gender_group["Price"].sum(),2)
#Avg purchase price per person
avg_pur_price_per_gender = round(total_value_gender/gender_count,2)
#Table and formatting
purchasing_analysis_gender = pd.DataFrame({
    "Purchase Count": purchase_count, "Average Purchase Price": avg_pur_price_gender, 
    "Total Purchase Value": total_value_gender, "Average Total Purchase per Person":avg_pur_price_per_gender
})
purchasing_analysis_gender["Average Purchase Price"] = (
purchasing_analysis_gender["Average Purchase Price"].astype(float).map("${:,.2f}".format))
purchasing_analysis_gender["Total Purchase Value"] = (
purchasing_analysis_gender["Total Purchase Value"].astype(float).map("${:,.2f}".format))
purchasing_analysis_gender["Average Total Purchase per Person"] = (
purchasing_analysis_gender["Average Total Purchase per Person"].astype(float).map("${:,.2f}".format))
purchasing_analysis_gender
```
